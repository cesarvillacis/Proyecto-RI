```python
from sklearn.datasets import fetch_20newsgroups

newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
newsgroupsdocs = newsgroups.data
```

```python
import nltk
from nltk.tokenize import word_tokenize
import pandas as pd

nltk.download('punkt', quiet=True)

newsgroup_df = pd.DataFrame(newsgroupsdocs, columns=['document'])

newsgroup_df['tokens'] = newsgroup_df['document'].apply(lambda x: word_tokenize(x, language='english'))
newsgroup_df['tokens'][0]

```

```python
from nltk import regexp_tokenize


newsgroup_df['regex_tokens'] = newsgroup_df['document'].str.lower().apply(regexp_tokenize, pattern=r'\w[a-z]+')
newsgroup_df
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

stopwords.words('english')
```

```python
def remove_stopwords(tokens):
  sw = stopwords.words('english')
  for w in sw:
    try:
      tokens.remove(w)
    except ValueError:
      pass
  return tokens
```

```python
newsgroup_df['sw_tokens'] = newsgroup_df['regex_tokens'].apply(remove_stopwords)
newsgroup_df
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer as wnl
```

```python
def lemmatize(tokens):
  return [wnl().lemmatize(t) for t in tokens]
```

```python
newsgroup_df['lem_tokens'] = newsgroup_df['sw_tokens'].apply(lemmatize)
newsgroup_df
```

```python
newsgroup_df['prep_doc'] = newsgroup_df['lem_tokens'].str.join(' ')
newsgroup_df
```
